```python
import pandas as pd        # For data manipulation
import seaborn as sns        # For data viz
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['movies.csv']))
```

```python
# Check for missing/NULL values

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

```python
# columns with missing values

[col for col in df.columns if df[col].isnull().any()]
```

```python
# Data types for our columns

df.dtypes
```

```python
# Create correct year column

df[['a','b']] = df['released'].str.split('(', expand=True) # split the string

df['date'] = pd.to_datetime(df['a'])        # string to date 

df['newdate']=df['date'].astype(str)        # date to string

df['correctyear'] = df['newdate'].astype(str).str[:4]    # Extracting year from date string
df.head()
```

```python
# order by gross column

df = df.sort_values(by=['gross'],inplace=False, ascending=False) 
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop duplicate

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# scatter plot between budget and gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget Vs Gross Earnings')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earning')

plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"purple"})
plt.title('Budget Vs Gross Earnings')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earning')
```

```python
# Correlation
# types: pearson, kendall, spearman
df.corr(method='pearson')
```

```python
df = df.sort_values(by=['gross'],inplace=False, ascending=False) 
df.head()
```

```python
# correlation is 0.74 

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numerous Features')

plt.xlabel('Movie features')

plt.ylabel('Movie features')

plt.show()
```

```python
# Look at Company

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()     
```

```python
df = df.sort_values(by=['gross'],inplace=False, ascending=False) 
df.head()
```

```python
# correlation is 0.74 

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numerous Features')

plt.xlabel('Movie features')

plt.ylabel('Movie features')

plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr.head()
```
